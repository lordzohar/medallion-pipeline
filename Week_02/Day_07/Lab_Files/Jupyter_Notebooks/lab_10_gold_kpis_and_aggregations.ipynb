## 0. Notebook Setup

Run this first. It moves the kernel into `Lab_Files` and makes the shared lab helpers importable.

```python
from pathlib import Path
import os
import sys
import types
import typing

# PySpark 3.4 imports typing.io, which is absent in Python 3.14+.
if "typing.io" not in sys.modules:
    typing_io = types.ModuleType("typing.io")
    typing_io.IO = typing.IO
    typing_io.TextIO = typing.TextIO
    typing_io.BinaryIO = typing.BinaryIO
    sys.modules["typing.io"] = typing_io

def find_lab_files(start: Path) -> Path:
    relative_options = [
        Path("."),
        Path("Lab_Files"),
        Path("Day_07") / "Lab_Files",
        Path("Week_02") / "Day_07" / "Lab_Files",
    ]
    for root in [start, *start.parents]:
        for relative in relative_options:
            candidate = (root / relative).resolve()
            if (candidate / "labs" / "day7_common.py").exists():
                return candidate
    raise FileNotFoundError(
        "Could not find Week_02/Day_07/Lab_Files. "
        "Start Jupyter from the repository root, Week_02/Day_07, or Week_02/Day_07/Lab_Files."
    )

HERE = Path.cwd().resolve()
LAB_FILES = find_lab_files(HERE)

os.chdir(LAB_FILES)
labs_path = str(LAB_FILES / "labs")
if labs_path not in sys.path:
    sys.path.insert(0, labs_path)

print(f"Working directory: {Path.cwd()}")
print(f"Using lab helpers from: {labs_path}")
```

## 1. Import Lab Helpers

```python

import importlib
import day7_common
day7_common = importlib.reload(day7_common)

from day7_common import LAKE_DIR, gold_frames, ensure_output_dirs, require_source_data, spark_session, write_csv_dir, read_parquet, write_parquet
```

## 2. Start Spark and Read Enriched Orders

```python
require_source_data()
ensure_output_dirs()
spark = spark_session("Day7Notebook10GoldKpis")

enriched = read_parquet(spark, LAKE_DIR / "silver" / "orders_enriched")
enriched.select("order_id", "event_date", "status", "region", "loyalty_tier", "product_category", "amount_usd").show(20, truncate=False)
```

## 3. Build Gold Frames

```python
frames = gold_frames(enriched)
gold_path = LAKE_DIR / "gold"
print(sorted(frames.keys()))
```

## 4. Daily Revenue

```python
frames["daily_revenue"].show(20, truncate=False)
```

## 5. Category Revenue

```python
frames["category_revenue"].show(20, truncate=False)
```

## 6. Segment Revenue

```python
frames["segment_revenue"].show(20, truncate=False)
```

## 7. Write Gold Outputs

```python
for name, frame in frames.items():
    write_parquet(frame, gold_path / name, mode="overwrite")
    write_csv_dir(frame, gold_path / f"{name}_notebook_csv")
print(f"Gold tables written: {', '.join(frames.keys())}")
```
